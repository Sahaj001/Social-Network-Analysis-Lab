```
#@title Creating a function for different types of centrality
```

**Centrality measures:** 

  * Degree centrality :
    assigns an importance score by number of links held by each node.


  * Betweenness centrality : 
    number of times a node lies on the shortest path between other nodes.


  * Closeness centrality : 
    each node based on their ‘closeness’ to all other nodes in the network.


  * Eigenvector centrality : 
    extension of degree centra
    lity 


  * Edge betweenness centrality : 
    number of shortest path that go through a edge

```
def all_centrality(dataset_name):
  print(nx.degree_centrality(dataset_name))
  print(nx.betweenness_centrality(dataset_name))
  print(nx.closeness_centrality(dataset_name))
  print(nx.eigenvector_centrality(dataset_name))
  print(nx.edge_betweenness_centrality(dataset_name))
  return
```

```
#@title Printing all centrality value
```

```
all_centrality(G_dolphin)
all_centrality(G_football)
all_centrality(G_karate)
```

```
#@title Function for plotting network dataset
```

```
# color_pattern as in cmap, with_label for labelling nodes
def plot_graph(Graph,color_pattern='magma',with_label=False,title=''):

  cluster = nx.clustering(Graph)

  m=defaultdict(list)
  for a in cluster:
    m[cluster[a]]=(cluster[a])*9999999

  color = []
  for nodes in Graph:
    color.append(m[cluster[nodes]])

  ax = plt.gca()
  ax.set_title(str(title))
  nx.draw(Graph,node_color = color,cmap = plt.get_cmap(color_pattern),with_labels=with_label,ax=ax)
  _ = ax.axis('off')
  plt.show()
  return
```

```
plot_graph(G_dolphin,'magma',title = 'Dolphin Network')
plot_graph(G_football,'rainbow',title = 'Football Network')
plot_graph(G_karate,'viridis',with_label=True,title = 'Karate Network')
```
