# Experiment 2 : Generate following random graphs and analyze the pros and cons : 


1. Erdos-Renyi Model
2. Albert-Barabasi Model graphs 
3. LFR graphs 

```
# import all dependencies
import matplotlib.pyplot as plt
import networkx as nx
```

```
#@title Erdos-Renyi Model Graphs
```

```
G1 = nx.erdos_renyi_graph(n=100,p=0.05)
sequence_value1 = []
node_size = []
for nodes in G1:
  node_size.append(nodes*3)
  sequence_value1.append(nodes)


cmap1 = plt.get_cmap('magma')

nx.draw(G1, node_size=node_size,node_color=sequence_value1,cmap=cmap1)
plt.show()
```

```
#@title Barabasi-Albert Model Graphs
```

```

G2 = nx.barabasi_albert_graph(n=100,m=2)
cmap = plt.get_cmap('viridis')
sequence_value = []
for nodes in G2:
  sequence_value.append(nodes)

nx.draw_networkx(G2,pos=nx.spring_layout(G2),alpha=0.5,node_color=sequence_value,cmap=cmap)
plt.show()
```

```
#@title LFR graphs
```

```
G3 = nx.LFR_benchmark_graph(n=50,tau1=30,tau2=10,mu=0.7,min_degree=5)
sequence_value3 = []
for nodes in G3:
  sequence_value3.append(nodes)
nx.draw_networkx(G3,pos=nx.spring_layout(G3),cmap=plt.get_cmap('magma'),node_color=sequence_value3)
plt.show()
```
